```python
def containOnDict(element, _dict):
    exists = False
    for key in _dict:
        if(element in _dict[key]):
            exists = True
    return exists

def find(directory, tags, subTag):
    list_obj = {}
    for filename in os.listdir(directory):
        name = '.'.join(filename.split('.')[:-1])
        extension = filename.split('.')[-1]
        list_obj[name] = []
        if(extension == 'xml'):
            r = open(directory+filename, 'r')
            xml = r.read()
            soup = BeautifulSoup(xml)
            for tag in tags:
                for q in soup.find_all(tag):
                    element = ''
                    if(subTag):
                        element = q[subTag]
                    else:
                        element = q.text
                    if(not containOnDict(element.lower(), list_obj) and element != ''):
                        list_obj[name].append(element.lower())
            r.close()
        if(extension == 'txt'):
            r = open(directory+filename, 'r')
            txt = r.read()
            elements = txt.replace('\r', '').split('\n')
            for element in elements:
                if(not containOnDict(element.lower(), list_obj) and element != ''):
                    list_obj[name].append(element.lower())
            r.close()
        if(extension == 'csv'):
            df = pandas.read_csv(directory+filename)
            for tag in tags:
                try:
                    #print(df[tag])
                    for element in df[tag]:
                        if(not containOnDict(element.lower(), list_obj) and element != ''):
                            list_obj[name].append(element.lower())
                except:
                    continue
                    
    return list_obj
```

```python
def findQandA(directory, tags, subTag):
    list_obj = {}
    qa = pandas.DataFrame(columns=['QUESTION', 'ANSWER'])
    for filename in os.listdir(directory):
        name = '.'.join(filename.split('.')[:-1])
        extension = filename.split('.')[-1]
        list_obj[name] = []
        if(extension == 'xml'):
            r = open(directory+filename, 'r')
            xml = r.read()
            soup = BeautifulSoup(xml)
            for tag in tags:
                for q in soup.find_all(tag):
                    element = ''
                    if(subTag):
                        element = q[subTag]
                    else:
                        element = q.text
                    if(not containOnDict(element.lower(), list_obj) and element != ''):
                        list_obj[name].append(element.lower())
                        qa.loc[len(qa)] = [element, q.parent.a.text]
            r.close()
        if(extension == 'txt'):
            r = open(directory+filename, 'r')
            txt = r.read()
            elements = txt.replace('\r', '').split('\n')
            for element in elements:
                if(not containOnDict(element.lower(), list_obj) and element != ''):
                    list_obj[name].append(element.lower())
            r.close()
        if(extension == 'csv'):
            df = pandas.read_csv(directory+filename)
            for tag in tags:
                try:
                    #print(df[tag])
                    for element in df[tag]:
                        if(not containOnDict(element.lower(), list_obj) and element != ''):
                            list_obj[name].append(element.lower())
                    qa = qa.append(df[['QUESTION', 'ANSWER']], ignore_index=True)
                except:
                    continue
                    
    return qa
```
